# KNN Class

```python
import numpy as np
from collections import Counter

# Define a custom K-Nearest Neighbors (KNN) Classifier
class KNNClassifier:

    # Constructor to initialize the number of neighbors (k)
    def __init__(self, k):
        self.n_neighbors = k        # Number of nearest neighbors to consider
        self.X_train = None         # Placeholder for training features
        self.y_train = None         # Placeholder for training labels

    # Fit method to store the training data
    def fit(self, X_train, y_train):
        # Convert input training data to NumPy arrays for efficient computation
        self.X_train = np.array(X_train)
        self.y_train = np.array(y_train)

    # Method to calculate Euclidean distance between two points
    def calculate_distance(self, i, j):
        return np.linalg.norm(i - j)  # sqrt(sum((i - j)^2))

    # Method to determine the majority class label among nearest neighbors
    def majority_counts(self, neighbors):
        # Extract labels of the nearest neighbors using their indices
        votes = [self.y_train[idx] for idx, _ in neighbors]
        # Count the frequency of each class label
        count = Counter(votes)
        # Return the label with the highest count (most common)
        return count.most_common(1)[0][0]

    # Predict method to classify unseen test data
    def predict(self, X_test):
        X_test = np.array(X_test)  # Ensure test data is a NumPy array
        y_pred = []                # Initialize an empty list to store predictions

        # Iterate over each test instance
        for i in X_test:
            distances = []  # List to store distances from current test point to all training points

            # Calculate distance from test point to every training point
            for j in range(len(self.X_train)):
                dist = self.calculate_distance(i, self.X_train[j])
                distances.append((j, dist))  # Store as (index, distance) tuple

            # Sort distances by the second item in the tuple (actual distance)
            neighbors = sorted(distances, key=lambda x: x[1])[:self.n_neighbors]

            # Find the majority class among the nearest neighbors
            label = self.majority_counts(neighbors)

            # Append the predicted label to the output list
            y_pred.append(label)

        # Return the predicted labels as a NumPy array
        return np.array(y_pred)
```

```python
model = KNNClassifier(k= 5)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
