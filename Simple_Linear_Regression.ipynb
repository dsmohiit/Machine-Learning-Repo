```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/placement.csv")
```

```python
# Ploting the scatter plot.
# CGPA and Package have a sort of linear relationship.
plt.scatter(df["cgpa"], df["package"], alpha= 0.5)
plt.xlabel("CGPA")
plt.ylabel("PACKAGE")

plt.title("CGPA vs PACKAGE")
plt.show()
```

```python
x = df.iloc[:, 0:1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
predicted_ctc = lr.predict(x_test)
predicted_ctc = pd.DataFrame({
    "predicted_ctc": predicted_ctc
})

predicted_ctc
```

```python
plt.scatter(df["cgpa"], df["package"], alpha= 0.5)
plt.plot(x_train, lr.predict(x_train), color= "red", label= "Best Fit Line")
plt.xlabel("CGPA")
plt.ylabel("PACKAGE")

plt.title("CGPA vs PACKAGE")
plt.legend()
plt.show()
```

```python
# Value of slope (m).
m = lr.coef_
m
```

```python
# Value of intercept (b).
b = lr.intercept_
b
```

```python
class MyLinearRegression():

  def __init__(self):
    self.m = None
    self.b = None

  def fit(self, x_train, y_train):
    x_train = np.array(x_train).flatten()
    y_train = np.array(y_train).flatten()

    x_train_mean = x_train.mean()
    y_train_mean = y_train.mean()


    num = np.sum((x_train - x_train_mean) * (y_train - y_train_mean))
    den = np.sum((x_train - x_train_mean) ** 2)

    self.m = num / den
    self.b = y_train_mean - (self.m * x_train_mean)

  def predict(self, x_test):
    return (self.m * x_test) + self.b

  def coef_(self):
    return self.m

  def intercept_(self):
    return self.b
```

```python
mlr = MyLinearRegression()
```

```python
mlr.fit(x_train, y_train)
```

```python
mlr.coef_()
```

```python
mlr.predict(x_test)
```
