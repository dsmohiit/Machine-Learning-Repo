# PCA -- Class

```python
class MyPCA:
  def __init__(self, n_components):

    """
    Parameter:
      n_components: int
        No. of compenents to keep.

    Attribute:
      components_: nd.array
        Contains component (eigenvectors) sorted by their importance (eigenvalues).

      explain_variance_ratio_: float
        Percatage of variance explain by n_components

      mean: float
        Mean of each feature.

      scale_: float
        Standard deviation of each feature.
    """

    self.n_components = n_components
    self.components_ = None
    self.explain_variance_ratio_ = None
    self.mean = None
    self.scale_ = None


  def fit(self, X):

    # Standardize the data
    self.mean = np.mean(X, axis= 0)
    self.scale_ = np.std(X, axis= 0)

    X_std = (X - self.mean) / self.scale_

    # Calulate Covariance Matrix
    cov_matrix = np.cov(X_std.T)

    # Find eigenvectors and eigenvalues
    eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)

    # Sorting eigenvectors according to their values
    idx = np.argsort(eigen_values)[::-1]
    sorted_eigen_values = eigen_values[idx]
    sorted_eigen_vectors = eigen_vectors[:, idx]

    # Selecting n_components from sorted_eigen_vectors
    self.components_ = sorted_eigen_vectors[:, :self.n_components]

    # Total Variance
    total_variance = np.sum(eigen_values)

    # Variance explained by n_components
    self.explain_variance_ratio_ = sorted_eigen_values[: self.n_components] / total_variance

    return self

  def transform(self, X):

    if self.components_ is None:
      raise RuntimeError("You must call fit() before you can transform the data.")

    X_std = (X - self.mean) / self.scale_

    X_projected = np.dot(X_std, self.components_)

    return X_projected

  def fit_transform(self, X):

    self.fit(X)

    return self.transform(X)
```
